```python
### at least once a month

file_path = r'C:\Users\Lenovo\Desktop\modaam\excel core/orders_pezeshki_noerror.csv' 
df = pd.read_csv(file_path)
                 #, parse_dates=['pay_at'], date_parser=lambda x: pd.to_datetime(x, format='%m/%d/%Y %I:%M:%S %p'))


# Assuming df is your DataFrame containing the data
# Convert 'pay_at' to datetime if it's not already in datetime format
df['pay_at'] = pd.to_datetime(df['pay_at'])

# Extract month and year for grouping
df['month_year'] = df['pay_at'].dt.to_period('M')

# Group by 'card_number' and 'month_year', count the occurrences
customer_frequency = df.groupby(['card_number', 'month_year']).size().reset_index(name='frequency')

# Filter customers who bought at least once a month
monthly_customers = customer_frequency.groupby('card_number').filter(lambda x: len(x) >= 1)['card_number'].unique()

filtered_df = df[df['card_number'].isin(monthly_customers)]
filtered_df
```

```python
### twice or more in month
file_path = r'C:\Users\Lenovo\Desktop\modaam\excel core/orders_pezeshki_noerror.csv' 
df = pd.read_csv(file_path)

# Assuming your data is stored in a DataFrame named df
# Convert 'pay_at' column to datetime if it's not already in datetime format
df['pay_at'] = pd.to_datetime(df['pay_at'])

# Group by 'card_number' and extract the month-year from 'pay_at' to count purchases per month per customer
df['month_year'] = df['pay_at'].dt.to_period('M')  # Extract month and year

# Count the number of purchases (transactions) per customer per month
customer_purchase_counts = df.groupby(['card_number', 'month_year']).size().reset_index(name='purchase_count')

# Filter customers who have purchased at least twice a month (or more)
frequent_buyers = customer_purchase_counts[customer_purchase_counts['purchase_count'] >= 2]

# Get the list of customers who meet the criteria
frequent_buyers_list = frequent_buyers['card_number'].unique()
filtered_df = df[df['card_number'].isin(frequent_buyers_list)]
filtered_df
```

```python
file_path = r'C:\Users\Lenovo\Desktop\modaam\excel core/orders_pezeshki_noerror.csv' 
df = pd.read_csv(file_path)
                 #, parse_dates=['pay_at'], date_parser=lambda x: pd.to_datetime(x, format='%m/%d/%Y %I:%M:%S %p'))

# Assuming your dataframe is named df
# Convert 'pay_at' column to datetime
df['pay_at'] = pd.to_datetime(df['pay_at'])

# Extract month and year into separate columns
df['month'] = df['pay_at'].dt.month
df['year'] = df['pay_at'].dt.year

# Group by 'card_number', 'month', and 'year', count transactions
user_monthly_counts = df.groupby(['card_number', 'month', 'year']).size().reset_index(name='buying_frequency')

# Drop the month and year columns if you don't need them anymore
df.drop(['month', 'year'], axis=1, inplace=True)

# Merge back to the original dataframe
df = pd.merge(df, user_monthly_counts, on='card_number', how='left')

df
```
