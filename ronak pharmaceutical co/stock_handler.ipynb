```python
import pandas as pd
import numpy as np
import os
def storck_handler(stock):
    diff = np.diff(stock.to_numpy(),1)
    diff = diff[diff>0].sum()
    return stock.to_numpy()[0]  + diff
```

```python
mapping = pd.read_excel("AllDistCenterProvince.xlsx")
stock = []
for path in os.listdir("stock"):
    df = pd.read_excel(f"stock/{path}")
    stock.append(df)
stock = pd.concat(stock,axis=0)
stock.rename(columns={"distcenternew":"DistributionCenter"},inplace=True)
dm = stock.merge(mapping, on=["Distributor","DistributionCenter"])

dm = dm.groupby(["Line","Month","ProductNameEng","Province","day"]).agg({"StockVolume":sum}).reset_index()
dm =  dm.groupby(["Line","Month","ProductNameEng","Province"]).agg({"StockVolume":storck_handler}).reset_index()
# dm.to_excel("stock_sale.xlsx")
target = []
for path in os.listdir("target"):
    df = pd.read_excel(f"target/{path}", index_col=0)
    target.append(df)
df_target = pd.concat(target, axis=0)
dm = dm.merge(df_target, on=["Line","ProductNameEng", "Month", "Province"])
dm["has_stock"] = dm["StockVolume"]>= dm["province_target"]*2
dm.to_excel("sale_stock.xlsx",index=False)
```
